```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
batch_size = 1
image_size = 256
num_workers = 2
preprocess = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.0, 0.0, 0.0],
         std=[1., 1., 1.]
         ),
])
```

```python
class FacadesImageFolder(torchvision.datasets.ImageFolder):
    def __getitem__(self, index):
        _chunk_num=2
        _dim = 2
        _path, _ = self.samples[index]
        X = self.loader(_path)
        if self.transform:
            X = self.transform(X)
        return torch.chunk(X, chunks=_chunk_num, dim=_dim)
```

```python
train_data = FacadesImageFolder(root=os.path.join(path, "train"), transform=preprocess)

train_loader = DataLoader(
    train_data,
     batch_size=batch_size,
      shuffle=True,
       num_workers=num_workers
       )
```

```python
class ContrUnetBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=4, stride=2,padding=1, with_norm=False, with_droput=False, with_maxpool=False):
        super().__init__()
        sequence = [nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, stride=stride, padding=padding)]
        if with_norm:
          sequence.append(nn.BatchNorm2d(out_channels))
        if with_maxpool:
          sequence.append(nn.MaxPool2d((2, 2)))
        if with_droput:
          sequence.append(nn.Dropout2d(0.5))
        sequence.append(nn.LeakyReLU(0.2))
        self.convs = nn.Sequential(*sequence)

    def forward(self, x):
        out = self.convs(x)
        return out
      
```

```python
class ExpUnetBlock(nn.Module):
    def __init__(self, in_channels, out_channels, with_norm=False, with_dropout=False):
        super().__init__()
        sequence = [nn.ConvTranspose2d(in_channels, out_channels, kernel_size=4, stride=2, padding=1), nn.ReLU()]
        if with_norm:
          sequence.append(nn.BatchNorm2d(out_channels))
        if with_dropout:
          sequence.append(nn.Dropout2d(0.5))
        self.deconvs = nn.Sequential(*sequence)
        
    def forward(self, x):
        return self.deconvs(x) 
```

```python
class Unet(nn.Module):
    def __init__(self, n_base_channels=64):
        super().__init__()
        self.contrast_blocks = nn.ModuleList([
            ContrUnetBlock(3, n_base_channels),
            ContrUnetBlock(n_base_channels, n_base_channels * 2, with_norm=True),
            ContrUnetBlock(n_base_channels * 2, n_base_channels * 4, with_norm=True),
            ContrUnetBlock(n_base_channels * 4, n_base_channels * 8, with_norm=True),
            ContrUnetBlock(n_base_channels * 8, n_base_channels * 8, with_norm=True, with_droput=True),
            ContrUnetBlock(n_base_channels * 8, n_base_channels * 8, with_norm=True),
            ContrUnetBlock(n_base_channels * 8, n_base_channels * 8, with_droput=True),
        ])
        self.expand_blocks = nn.ModuleList([
            ExpUnetBlock(n_base_channels * 8, n_base_channels * 8, with_norm=False, with_dropout=True),
            ExpUnetBlock(n_base_channels * 16, n_base_channels * 8, with_norm=True, with_dropout=True),
            ExpUnetBlock(n_base_channels * 16, n_base_channels * 8, with_norm=True),
            ExpUnetBlock(n_base_channels * 16, n_base_channels * 4, with_norm=True),
            ExpUnetBlock(n_base_channels * 8, n_base_channels * 2, with_norm=True),
            ExpUnetBlock(n_base_channels * 4, n_base_channels * 1, with_norm=False),
        ])
        self.final_block = nn.ConvTranspose2d(n_base_channels * 1, 3, kernel_size=4, stride=2, padding=1)
            
        
    def forward(self, x):
        out = x
        outputs = []
        for i, block in enumerate(self.contrast_blocks):
            out = block(out)
            outputs.append(out)
        for i, block in enumerate(self.expand_blocks[:-1]):   
            out = block(out)
            out = torch.cat((out, outputs[-i - 2]), axis=1)

        out = self.expand_blocks[-1](out)
        out = self.final_block(out)
        return out
```

```python
class Discriminator(nn.Module):
    def __init__(self, in_channels=3, n_base_channels=64, kernel_size=4, strides=2, padding=1):
        super().__init__()
        self.blocks = nn.ModuleList([
            ContrUnetBlock(in_channels * 2, n_base_channels, kernel_size, strides, padding),
            ContrUnetBlock(n_base_channels * 1, n_base_channels * 2, kernel_size, strides, padding, with_norm=True),
            ContrUnetBlock(n_base_channels * 2, n_base_channels * 4, kernel_size, strides, 1, with_norm=True),
            ContrUnetBlock(n_base_channels * 4, n_base_channels * 8, kernel_size, strides//2 , 1, with_norm=True),
            ContrUnetBlock(n_base_channels * 8, 1, kernel_size, strides//2, 0)
        ])
        self.sigmoid = nn.Sigmoid()

    def forward(self, x1, x2):
        x = torch.cat([x1, x2], axis=1)
        for i, block in enumerate(self.blocks):
            x = block(x)
        return self.sigmoid(x)
```

```python
def weights_init(m):
  if isinstance(m, nn.Conv2d) or isinstance(m, nn.ConvTranspose2d):
    torch.nn.init.xavier_normal_(m.weight.data)
```

```python
### Part1
```

```python
generator = Unet(n_base_channels=64).to(device)
discriminator = Discriminator().to(device)
```

```python
generator.apply(weights_init)
discriminator.apply(weights_init)
```

```python
beta1, beta2 = 0.5, 0.999
generator_optim = torch.optim.Adam(generator.parameters(), lr=2e-4, betas=(beta1, beta2))
discriminator_optim = torch.optim.Adam(discriminator.parameters(), lr=2e-4, betas=(beta1, beta2))
```

```python
crit = nn.BCEWithLogitsLoss()
crit_recon = nn.L1Loss()
```

```python
def discr_loss_func(discr, real_val, fake_val, x):
    pred_real = discr(real_val, x)
    pred_fake = discr(fake_val, x)
    return crit(
        pred_real,
        torch.ones_like(pred_real)) + crit(pred_fake, torch.zeros_like(pred_fake)) / 2
```

```python
def gener_loss_func(pred_fake):
    return crit(pred_fake, torch.ones_like(pred_fake))
```

```python
def discriminator_train(discriminator, generator, real_val, x):
    fake_val = generator(x)
    discriminator_optim.zero_grad()
    loss = discr_loss_func(discriminator, real_val, fake_val, x)
    loss.backward()
    discriminator_optim.step()
    return loss.item()
```

```python
def generator_train(discriminator, generator,real_val, x):
    fake_val = generator(x)
    generator_optim.zero_grad()
    fake_pred = discriminator(fake_val, x)
    loss = gener_loss_func(fake_pred)
    loss_rec = crit_recon(real_val, fake_val) 
    loss += 100 * loss_rec
    loss.backward()
    generator_optim.step()
    return loss.item()
```

```python
loss_discr_data, loss_gen_data = list(), list()
```

```python
epoch_num = 18
val_path = path + "/val/imgs"
```

```python
def run():
    generator.train(True)
    discriminator.train(True)
    for x, y in tqdm(train_loader):
        y = y.to(device)
        x = x.to(device)
        real_x = x.clone()
        discr_loss = discriminator_train(discriminator, generator, real_x, y)
        gen_loss = generator_train(discriminator, generator, real_x, y)

    loss_discr_data.append(discr_loss)
    loss_gen_data.append(gen_loss)
    clear_output(True)
    generator.eval()
    discriminator.eval()
    with torch.no_grad():
        y = val_data[0][1]
        y = y.to(device).unsqueeze(0)
        img = generator(y)
        img = torch.clamp(img, min=0, max=1)
        plt.figure(figsize=(16,5))
        plt.subplot(1,3,1)
        plt.title("Gen, discr loss")
        plt.ylabel('Loss')
        plt.xlabel('EpochNum')
        last_discr_val, last_gen_val = loss_discr_data[-1], loss_gen_data[-1]
        plt.plot(loss_discr_data, label='Loss discr: {:.4f}'.format(last_discr_val))
        plt.plot(loss_gen_data, label='Loss gen: {:.4f}'.format(last_gen_val))
        plt.legend()
        plt.subplot(1,3,2)
        plt.title("FakeImg")
        plt.imshow(img.cpu().detach().numpy().transpose([0, 2, 3, 1])[0]) 
        plt.subplot(1,3,3)
        plt.title("ValImg")
        img = mpimg.imread(val_path + "/1.jpg")
        imgplot = plt.imshow(img)      
        plt.show()
```

```python
for i, epoch in enumerate(range(1, epoch_num)):
    print("Epoch num is:", i)
    run()
```

```python
### Part2
```

```python
generator_a = Unet(n_base_channels=64).to(device)
discriminator_a = Discriminator().to(device)

generator_b = Unet(n_base_channels=64).to(device)
discriminator_b = Discriminator().to(device)
```

```python
generator_a.apply(weights_init)
discriminator_a.apply(weights_init)

generator_b.apply(weights_init)
discriminator_b.apply(weights_init)
```

```python
crit = nn.BCEWithLogitsLoss()
crit_recon = nn.L1Loss()
```

```python
beta1, beta2 = 0.5, 0.999

generator_optim_a = torch.optim.Adam(generator_a.parameters(), lr=2e-4, betas=(beta1, beta2))
generator_optim_b = torch.optim.Adam(generator_b.parameters(), lr=2e-4, betas=(beta1, beta2))

discriminator_optim_a = torch.optim.Adam(discriminator_a.parameters(), lr=2e-4, betas=(beta1, beta2))
discriminator_optim_b = torch.optim.Adam(discriminator_b.parameters(), lr=2e-4, betas=(beta1, beta2))
```

```python
def discr_loss_func(discr, real_val, fake_val, x):
    pred_real = discr(real_val, x)
    pred_fake = discr(fake_val, x)
    return crit(pred_real, torch.ones_like(pred_real)) + crit(pred_fake, torch.zeros_like(pred_fake)) / 2

def gener_loss_func(pred_fake):
    return crit(pred_fake, torch.ones_like(pred_fake))
```

```python
def discriminator_train(discriminator, generator, real_val, x, discriminator_optim):
    fake_val = generator(x)
    discriminator_optim.zero_grad()
    loss = discr_loss_func(discriminator, real_val, fake_val, x)
    loss.backward()
    discriminator_optim.step()
    return loss.item()
```

```python
def generator_train(discriminator_a, discriminator_b, generator_a, generator_b, x, y):
    fake_x = generator_a(y)
    rec_x = generator_b(fake_x)
    fake_y = generator_b(x)
    rec_y = generator_a(fake_y)
    generator_optim_a.zero_grad()
    generator_optim_b.zero_grad()
    pred_fake_x = discriminator_a(fake_x, x)
    pred_fake_y = discriminator_b(fake_y, y)
    loss = gener_loss_func(pred_fake_x) + gener_loss_func(pred_fake_y)  
    loss_recon = 5 * crit_recon(x, rec_x) + 5 * crit_recon(y, rec_y) + 10 * crit_recon(generator_b(y), y) + 10 * crit_recon(generator_a(x), x)
    loss += loss_recon
    loss.backward()
    generator_optim_a.step()
    generator_optim_b.step()
    return loss.item()
```

```python
loss_discr_data_a = list()
loss_discr_data_b = list()
loss_gen_data = list()
```

```python
epoch_num = 25
val_path = path + "/val/imgs"
```

```python
def run():
    generator_a.train(True)
    discriminator_b.train(True)
    generator_b.train(True)
    discriminator_b.train(True)
    for x, y in tqdm(train_loader):
        y = y.to(device)
        x = x.to(device)
        real_x = x.clone()
        real_y = y.clone()
        discr_loss_a = discriminator_train(discriminator_a, generator_a, real_x, y, discriminator_optim_a)
        discr_loss_b = discriminator_train(discriminator_b, generator_b, real_y, x, discriminator_optim_b)
        gen_loss = generator_train(discriminator_a, discriminator_b, generator_a, generator_b, x, y)

    loss_discr_data_a.append(discr_loss_a)
    loss_gen_data.append(gen_loss)
    loss_discr_data_b.append(discr_loss_b)
    clear_output(True)
    generator_a.eval()
    discriminator_a.eval()
    generator_b.eval()
    discriminator_b.eval()

    with torch.no_grad():
        y = val_data[0][1]
        y = y.to(device).unsqueeze(0)
        img = generator_a(y)
        img = torch.clamp(img, min=0, max=1)
        plt.figure(figsize=(16,5))
        plt.subplot(1,3,1)
        plt.ylabel('Loss')
        plt.xlabel('EpochNum')
        plt.plot(loss_discr_data_a, label='Loss discr a: {:.4f}'.format(loss_discr_data_a[-1]))
        plt.plot(loss_discr_data_b, label='Loss discr b: {:.4f}'.format(loss_discr_data_b[-1]))
        plt.plot(loss_gen_data, label='Loss gen: {:.4f}'.format(loss_gen_data[-1]))
        plt.legend()
        plt.subplot(1,3,2)
        plt.title("FakeImg")
        plt.imshow(img.cpu().detach().numpy().transpose([0, 2, 3, 1])[0])
        plt.subplot(1,3,3)
        plt.title("ValImg")
        img = mpimg.imread(val_path + "/1.jpg")
        imgplot = plt.imshow(img)
        plt.show()
```

```python
for i, epoch in enumerate(range(1, epoch_num)):
    print("Epoch num is:", i)
    run()
```
